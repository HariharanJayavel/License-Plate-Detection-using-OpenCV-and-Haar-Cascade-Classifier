```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('input.jpg')
```

```python
def display(img):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax.imshow(new_img)
```

```python
display(img)
```

```python
plate_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_russian_plate_number.xml')
```

```python
def detect_plates(img):
    plate_img = img.copy()
    plate_rects = plate_cascade.detectMultiScale(plate_img,scaleFactor=1.1,minNeighbors=4,minSize=(30, 30))
    print(f"Total license plates detected: {len(plate_rects)}")
    for (x, y, w, h) in plate_rects:
        cv2.rectangle(plate_img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    return plate_img
```

```python
result = detect_plates(img)
```

```python
display(result)
```
